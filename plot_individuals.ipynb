```python
import pandas as pd

#df = pandas.read_csv('hf_no_int.csv')
df1 = pd.read_csv('chandra_no_int.csv')
df2 = pd.read_csv('chandra_int.csv')
df1_new = df1.dropna(axis = 0, how ='any') 
df2_new = df2.dropna(axis = 0, how ='any')

```

```python
df1_sorted=df1_new.sort_values(by=['dBratio'])
df2_sorted=df2_new.sort_values(by=['dBratio'])
```

```python
for index, rows in df1_sorted.iterrows():
    if (rows['Choice'])=='L':
        df1_sorted.Choice[index]=1
    if (rows['Choice'])=='Y':
        df1_sorted.Choice[index]=2
        
for index, rows in df2_sorted.iterrows():
    if (rows['Choice'])=='L':
        df2_sorted.Choice[index]=1
    if (rows['Choice'])=='Y':
        df2_sorted.Choice[index]=2
```

```python
df1_sorted.plot.hist(x='dBratio', y='Choice',alpha=0.7, title = " Laurel (Left) Vs Yanni (Right)")
col = df1_sorted.Choice.map({1:'b', 2:'r'})

df1_sorted.plot.scatter(x='dBratio', y='Confidence', c=col,alpha=0.4 , ylim=(0,110))
```

```python
df2_sorted.plot.hist(x='dBratio', y='Choice',alpha=0.7, title = " Laurel (Left) Vs Yanni (Right)")
col = df2_sorted.Choice.map({1:'b', 2:'r'})

df2_sorted.plot.scatter(x='dBratio', y='Confidence', c=col,alpha=0.4 , ylim=(0,110))
```

```python
df_newer = df1_sorted.set_index('dBratio').join(df2_sorted.set_index('dBratio'),lsuffix='_filter', rsuffix='_filter_int')
```

```python
df_newer=df_newer.reset_index()
```
